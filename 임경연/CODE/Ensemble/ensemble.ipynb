```python
import os
import pandas as pd
from collections import defaultdict
```

```python
PATH = "/opt/ml/level2-movie-recommendation-level2-recsys-15/임경연/CODE/Ensemble/DATA"
```

```python
weight = [10, 7, 6, 5, 4]

data_1 = pd.read_csv(os.path.join(PATH, "output (6).csv"))
data_2 = pd.read_csv(os.path.join(PATH, "output (2).csv"))
data_3 = pd.read_csv(os.path.join(PATH, "output (3).csv"))
data_4 = pd.read_csv(os.path.join(PATH, "output (4).csv"))
data_5 = pd.read_csv(os.path.join(PATH, "output (5).csv"))
```

```python
score = defaultdict(list)

data_1_ = data_1.values.tolist()
for u, i in data_1_:
    token = 0
    for j in score[u]:
        if i == j[0]:
            token = 1
            j[1] += 10
            break
    if token == 0:
        score[u].append([i, 10])

data_2_ = data_2.values.tolist()
for u, i in data_2_:
    token = 0
    for j in score[u]:
        if i == j[0]:
            token = 1
            j[1] += 7
            break
    if token == 0:
        score[u].append([i, 7])

data_3_ = data_3.values.tolist()
for u, i in data_3_:
    token = 0
    for j in score[u]:
        if i == j[0]:
            token = 1
            j[1] += 6
            break
    if token == 0:
        score[u].append([i, 6])

data_4_ = data_4.values.tolist()
for u, i in data_4_:
    token = 0
    for j in score[u]:
        if i == j[0]:
            token = 1
            j[1] += 5
            break
    if token == 0:
        score[u].append([i, 5])

# data_5_ = data_5.values.tolist()
# for u, i in data_5_:
#     token = 0
#     for j in score[u]:
#         if i == j[0]:
#             token = 1
#             j[1] += 4
#             break
#     if token == 0:
#         score[u].append([i, 4])
```

```python
answer = []
for i in score:
    score[i].sort(key = lambda x: -x[1])
    for j in range(10):
        answer.append([i, score[i][j][0]])
```
